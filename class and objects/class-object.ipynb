```python
class order:
    def __init__(self,num,t,c):
        self.number=num
        self.type=t
        self.amount=c 
    
    def set_ord_amt(self,n):
        self.amount=n
    def set_ord_type(self,t):
        self.type=t
    def set_ord_num(self,num):
        self.number=num
    def get_ord_amt(self):
        return self.amount
    def get_ord_type(self):
        return self.type
    def get_ord_num(self):
        return self.number
   
```

```python
def saveintofile(filename,l):
    with open(filename,"w") as file:
        for each in l:
            data=str(each.number)+","+each.type+","+str(each.amount)
            file.write(data)
    file.close()
        
    
```

```python
l=[]
ch=0
onum=0
while(ch!=6):
    print("==============Welcome to Pizza Palace - Class Demo==============")
    print("1. Display Orders.")
    print("2. Add New Order.")
    print("3. Remove Order.")
    print("4. Edit Order.")
    print("5. Save Order.")
    print("6. Exit Program.")
    ch=int(input("Selection> "))
    if(ch==1):
        print("** Order List **")
        for each in l:
            print("Order Number:",each.get_ord_num())
            print("Order Type:",each.get_ord_type())
            print("Order Amount:",each.get_ord_amt())
            print("---------")
        print("******************")
    if(ch==2):
        t=input("Enter Type: ")
        c=len(t)*5+100
        obj=order(onum,t,c)
        l.append(obj)
        print("Order Added")
        onum+=1
    if(ch==3):
        i=int(input("Enter Order Number:"))
        l.pop(i)
        print("Order Removed")
    if(ch==4):
        var=int(input("1. Change Order Number. \n2. Change Order Type.\n3. Change Order Amount.\n"))
        value=input()
        if(var==1):
            l[-1].set_ord_num(int(value))
        elif(var==2):
            l[-1].set_ord_type(value)
        else:
            l[-1].set_ord_amt(int(value))
        
    if(ch==5):
        filename=input("Enter file name: ")
        saveintofile(filename,l)
        print("Data Saved.")
        
print("Exiting Program...")

    

    
```
