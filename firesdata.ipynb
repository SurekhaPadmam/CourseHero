```python
import numpy as np
import pandas as pd
```

```python
fires_by_month=pd.read_pickle(r"C:\Users\Surekha\Downloads\fires_by_month.pkl")
print(fires_by_month.head())
```

```python
print(fires_by_month.info())
```

```python
# group the data by state and year
df=fires_by_month.groupby(['state','fire_year'])
print(df.head())
```

```python
print(df.first())
```

```python
# sum the grouped data and store DataFrame in fires_by_year variable
fires_by_year=df.sum()
print(fires_by_year)
```

```python
fires_by_year=fires_by_year.drop(columns=['fire_month'])
```

```python
print(fires_by_year)
```

```python
# select rows where the year is 2013 or later
fires2013 = fires_by_year[fires_by_year.index.get_level_values(1)>=2013]
print(fires2013)
print(fires2013.info())
```

```python
# use the pivot() method to pivot the data so the state values are used as row labels, 
# fire_year values are used as column labels, and acres_burned values are used as data
fires2013=fires2013.reset_index()
print(fires2013.info())
print(fires2013.pivot(index='state',columns='fire_year',values='acres_burned'))
```

```python
# use the pivot_table() method with the fires_by_month DataFrame to get the same result
# as the previous cell
print(fires_by_month.info())
```

```python
fires2013month=fires_by_month.pivot_table(index='state',columns='fire_year',values='acres_burned',aggfunc=np.sum)
print(fires2013month)
```

```python
print(fires2013month[[2013,2014,2015]])
```

## Work with bins

```python
# reset the index for the DataFrame named fires_by_year
print(fires_by_year.info())
```

```python
fires_by_year=fires_by_year.reset_index()
```

```python
# bin the data by decade
# double-check the edge values
fires_by_year['decade'] = pd.cut(fires_by_year.fire_year, 

                                 bins=[1990,2000,2010,2020], 

                                 labels=['1990s','2000s','2010s'], 

                                 right=False)
print(fires_by_year)
```

```python
# drop the fire_year column and assign the DataFrame that's returned to a variable named fires_by_decade
fires_by_decade=fires_by_year.drop(columns=['fire_year'])
print(fires_by_decade)
```

```python
# group by the state and decade columns and sum the data
final=fires_by_decade.groupby(['state','decade']).sum()
print(final)
```
